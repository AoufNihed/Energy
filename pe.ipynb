###  regression analysis 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#### Linear Regression Model (Energy Production as a Function of Injected Power)

```python
X_linear = df[['Injected_Power_MW']]
y = df['Energy_Produced_MWh']

X_train, X_test, y_train, y_test = train_test_split(X_linear, y, test_size=0.2, random_state=42)
```

```python
linear_model = LinearRegression()


linear_model.fit(X_train, y_train)


a = linear_model.coef_[0]
b = linear_model.intercept_

print(f'Linear Model: Energy_Produced_MWh = {a} * Injected_Power_MW + {b}')
```

```python
y_pred_linear = linear_model.predict(X_test)

#evaluation metrics
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)

print(f'Linear Model MSE: {mse_linear}')
print(f'Linear Model R-squared: {r2_linear}')
```

####  Multiple Regression Model (Incorporating Demand)

```python
X_multiple = df[['Injected_Power_MW', 'Real_Time_Demand_MW']]
y = df['Energy_Produced_MWh']


X_train, X_test, y_train, y_test = train_test_split(X_multiple, y, test_size=0.2, random_state=42)
```

```python
multiple_model = LinearRegression()


multiple_model.fit(X_train, y_train)

c, d = multiple_model.coef_
e = multiple_model.intercept_

print(f'Multiple Regression Model: Energy_Produced_MWh = {c} * Injected_Power_MW + {d} * Real_Time_Demand_MW + {e}')
```

```python
y_pred_multiple = multiple_model.predict(X_test)

# evaluation metrics
mse_multiple = mean_squared_error(y_test, y_pred_multiple)
r2_multiple = r2_score(y_test, y_pred_multiple)

print(f'Multiple Regression Model MSE: {mse_multiple}')
print(f'Multiple Regression Model R-squared: {r2_multiple}')
```

#### LSTM model for predicting the next day's energy consumption

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
```

```python
df.columns
```

```python
data = df[['Energy_Produced_MWh', 'Real_Time_Demand_MW']]  
```

```python
# Scale the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Create sequences for LSTM
sequence_length = 24  # time steps for each prediction
X, y = [], []
```

```python


# Data Preprocessing
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df[['Energy_Produced_MWh']])

# Define sequence length and prepare training data
sequence_length = 90
X, y = [], []
for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i-sequence_length:i, 0])  
    y.append(scaled_data[i, 0])  # Predict next step

X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))  # Reshape for LSTM input

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=100, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(LSTM(units=100))
model.add(Dense(units=1))

# Compile and train the model
model.compile(optimizer=Adam(learning_rate=0.0005), loss='mean_squared_error')
model.fit(X, y, epochs=50, batch_size=32)

# Make predictions
predictions = model.predict(X)

# Inverse transform predictions to original scale
predictions_with_dummy = np.concatenate([np.zeros((predictions.shape[0], 1)), predictions], axis=1)
predicted_demand_full = scaler.inverse_transform(predictions_with_dummy)
Energy_Produced_MWh = predicted_demand_full[:, 1]  # Extract demand column

# Calculate gas and steam production based on ratios
gas_ratio, steam_ratio = 0.6, 0.4
predicted_gas = Energy_Produced_MWh * gas_ratio
predicted_steam = Energy_Produced_MWh * steam_ratio

# Output results
results = pd.DataFrame({
    'Predicted_Demand': Energy_Produced_MWh,
    'Predicted_Gas': predicted_gas,
    'Predicted_Steam': predicted_steam
})

# Display first few rows
print(results.head())
```
